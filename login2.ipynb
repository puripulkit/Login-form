```python
import os,sys,re
def signup():
    regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    userlen=20
    temp=input("Enter username")
    if(re.fullmatch(regex, temp)):
        print("Valid Email")
 
    else:
        print("Invalid Email")
        return
    l=len(temp)
    temp=temp+((userlen-l)*"")
    temp=temp.encode()
    with open("userinfo.txt","ab+") as f:
        size=os.path.getsize("userinfo.txt")
        n=int(size/28)
        position=0
        for i in range (n):
            f.seek(position)
            str=f.read(20)
            if temp in str:
                print("username already exist")
                return
            position=position+28
    username=temp
    print("Enter Password")
    print("Password must contain one capital and one special character and must be in between 5 to 16 characters")
    u=0
    s=0
    a=input()
   # if(len(a)<=16 or len(a)>=5):
    for i in range(0,len(a)):
        if(a[i].isupper()):
            u=u+1
        elif(a[i]=="@"or a[i]=="$"or a[i]=="!"or a[i]=="#"or a[i]=="%"):
            s=s+1
    try:
        if(u!=0 and s!=0 and (len(a)<=16 or len(a)>=5)):
            password=a
            password=password.encode()
        else:
            raise Exception("invalid password")
    except Exception as e:
        print("Password must contain one capital and one special character and must be in between 5 to 16 characters")
        exit()
    with open("userinfo.txt","a+b") as f:
        f.write(username+password)
        
def login():
    regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    userlen=20
    user=input("Enter username")
    if(re.fullmatch(regex, user)):
        print("Valid Email")
 
    else:
        print("Invalid Email")
        return
    l=len(user)
    user=user+((userlen-l)*"")
    user=user.encode()
    password=input("Enter Password")
    password=password.encode()
    userinfo=user+password
    size=os.path.getsize("userinfo.txt")
    n=int(size/28)
    with open('userinfo.txt','rb') as f:
        position=0
        for i in range(n):
            f.seek(position)
            str=f.read(28)
            if userinfo in str:
                print("Login Successful")
                exit()
            position+=28
        print("username and password does not match")
        return
   
while(1):
    print("1.Signup\n2.Login\nExit")
    print("Enter your choice:")
    c=int(input())
    if(c==1):
        signup()
    elif(c==2):
        login()
    elif(c==3):
        exit()
    else:
        print("invalid choice")
```
